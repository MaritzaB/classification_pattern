# Teorema de Bayes

$P \left(\color{#6666ff}{A}  | \color{#ff6666}{B} \right)  = \displaystyle \frac  {p \left( \color{#ff6666}{B} | \color{#6666ff}{A} \right)   P \left(\color{#6666ff}{A}\right)}   {p \left(\color{#ff6666}{B}\right)} $ <br><br>

$ P\left(\color{#6666ff}{A} | \color{#ff6666}{B} \right) $: Probabilidad _a posteriori_ de que suceda $\color{#6666ff}{A}$, dado que sucedió $\color{#ff6666}{B}$.<br>

$ P\left(\color{#6666ff}{A}\right) $: Probabilidad _a priori_ de que suceda $\color{#6666ff}{A}$.<br>

$ p\left(\color{#ff6666}{B}  | \color{#6666ff}{A} \right) $: _Verosimilitud_ (_likelihood_). Probabilidad condicional de que suceda $\color{#ff6666}{B}$, dado que ocurrió $\color{#6666ff}{A}$.
<br>

$ p\left(\color{#ff6666}{B}\right) $: Evidencia de probabilidad de que suceda $\color{#ff6666}{B}$. Sirve como una constante de normalización.
<br>

<font size=1> Se muestran con **P** las probabilidades, y con **p** las funciones de densidad de probabilidad.</font>


# Problema de Chabelo y la catafixia

1.   En un concurso, el participante debe elegir una de tres puertas y se llevará el premio que esté detrás. En una puerta hay un auto, y en las otras dos, una cabra.

2.   Cuando el participante eligió la puerta, el presentador abre una de las dos puertas restantes, en la que resulta que hay una cabra.

3.   El presentador da opción al participante de mantener la puerta de su elección inicial, o bien, cambiar de puerta con la finalidad de ganar el auto.




image.png

¿Es mejor para el participante cambiar su elección o mantener la puerta inicial?

Supongamos que el concursante eligió la puerta número 3. ¿Cómo nos puede ayudar el Teorema de Bayes?

## Problema de Monty Hall con Teorema de Bayes

Definamos los eventos:<br>
$W_1$: que el premio esté en la puerta 1.<br>
$W_2$: que el premio esté en la puerta 2.<br>
$W_3$: que el premio esté en la puerta 3.<br>
<br>

---

<br>Las probabilidades *a priori* de estos eventos serían:
<br><br>
$P(W_1) = P(W_2) = P(W_3) = \displaystyle\frac{1}{3}$
<br><br>
Es decir, *a priori* hay un tercio de probabilidad de que el auto esté en alguna de las 3 puertas.
<br><br>

---

<br>El concursante ha elegido una puerta (digamos, la 3). Entonces el presentador abre una de las dos puertas restantes, que NO es la puerta del auto. Esto es:
<br><br>
$D_1$: el presentador abre la puerta 1.<br>
$D_2$: el presentador abre la puerta 2.<br>
$D_3$: el presentador abre la puerta 3.<br>
<br>

---

<br>Como el concursante eligió la puerta 3, el presentador abrirá alguna otra de las puertas. Las probabilidades de que abra una u otra puerta de las restantes (D1 o D2), dado que él sabe dónde está el premio, son: (*verosimilitudes*)
<br><br>
$ p(D_1 | W_1) = 0; p(D_2 | W_1) = 1 $<br>
$ p(D_1 | W_2) = 1; p(D_2 | W_2) = 0 $<br>
$ p(D_1 | W_3) = \frac{1}{2}; p(D_2 | W_3) = \frac{1}{2} $
<br><br>

---

<br>Por el teorema de la probabilidad total, es posible calcular la probabilidad de que el presentador abra cada puerta NO elegida por el concursante:
<br><br>
$ p(D_1) = P(W_1)·p(D_1|W_1) + P(W_2)·p(D_1|W_2) + P(W_3)·p(D_1|W_3) $
<br><br>
$ p(D_1) = \frac{1}{3}·0 + \frac{1}{3}·1 + \frac{1}{3}·\frac{1}{2} = \frac{1}{2} $
<br><br><br>
$ p(D_2) = P(W_1)·p(D_2|W_1) + P(W_2)·p(D_2|W_2) + P(W_3)·p(D_2|W_3) $
<br><br>
$ p(D_2) = \frac{1}{3}·1 + \frac{1}{3}·0 + \frac{1}{3}·\frac{1}{2} = \frac{1}{2} $

## Aplicando el Teorema de Bayes

*   Si el participante NO cambia de puerta, y el presentador abre la puerta 1 (D1).<br><br>
$ P(W_3|D_1) = \displaystyle\frac{p(D_1|W_3)·P(W_3)}{p(D_1)} $
<br><br>
$ P(W_3|D_1) = \displaystyle\frac{\frac{1}{2}·\frac{1}{3}}{\frac{1}{2}} = \frac{1}{3} $
<br>
<br>
*   Si el participante NO cambia de puerta, y el presentador abre la puerta 2 (D2).<br><br>
$ P(W_3|D_2) = \displaystyle\frac{p(D_2|W_3)·P(W_3)}{p(D_2)} $
<br><br>
$ P(W_3|D_2) = \displaystyle\frac{\frac{1}{2}·\frac{1}{3}}{\frac{1}{2}} = \frac{1}{3} $
<br>
<br>
Las probabilidades de ganar el auto, si el participante no cambia, permanecen en $\frac{1}{3}$, sin importar cuál sea la puerta que abra el presentador.
<br><br>

---

*   Si el participante SÍ cambia de puerta y el presentador abre la puerta 2 (D2).<br><br>
$ P(W_1|D_2) = \displaystyle\frac{p(D_2|W_1)·P(W_1)}{p(D_2)} $
<br><br>
$ P(W_1|D_2) = \displaystyle\frac{1·\frac{1}{3}}{\frac{1}{2}} = \frac{2}{3} $
<br>
<br>
*   Si el participante SÍ cambia de puerta y el presentador abre la puerta 1 (D1).<br><br>
$ P(W_2|D_1) = \displaystyle\frac{p(D_1|W_2)·P(W_2)}{p(D_1)} $
<br><br>
$ P(W_2|D_1) = \displaystyle\frac{1·\frac{1}{3}}{\frac{1}{2}} = \frac{2}{3} $
<br>
<br>
Las probabilidades de ganar el auto, si el participante decide cambiar de puerta, ascienden a $\frac{2}{3}$, sin importar cuál sea la puerta que abra el presentador.
<br><br>
<br>En conclusión: lo mejor es SIEMPRE cambiar de puerta.
